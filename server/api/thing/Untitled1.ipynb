```python
import pandas as pd
import numpy as np
import urllib
```

```python
data = pd.read_csv('expenses.csv')

```

```python
len (data.Receipt.dropna())
```

```python
data.columns
```

```python
set(data.Name)
```

```python
urlCodes = {'urlcode': []}
for i in data.Receipt:
    try:
        if i is not np.NAN and 'http' in i:
            urlCodes['urlcode'].append(urllib.urlopen(i).getcode())
        else:
            urlCodes['urlcode'].append(None)
    except IOError:
            urlCodes['urlcode'].append(None)
```

```python
len (urlCodes['urlcode']) == len(data)
```

```python
pd.concat([data, pd.DataFrame(urlCodes)], axis = 1).to_csv('cleaned_expenses.csv', index = False )
```
